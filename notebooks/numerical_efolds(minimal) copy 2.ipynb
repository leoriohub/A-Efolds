```python
from ipywidgets import interact, FloatSlider

# Constants (adjust these as needed)
λ = 1.0              # Lambda parameter
ξ = 0.1              # Xi parameter
M_p = 1.0            # Planck mass (subscript M_p)

def potential_U(h, v):
    """Calculate the potential U(h, v)"""
    numerator = h**2 - v**2
    denominator = 1 + (ξ * h**2) / M_p**2
    return (λ/4) * (numerator / denominator)**2

# Create the interactive plot
def plot_potential(v=0.5):
    h = np.linspace(-2, 2, 500)  # Range for h (adjust as needed)
    U = potential_U(h, v)
    
    plt.figure(figsize=(10, 6))
    plt.plot(h, U, linewidth=2, color='darkblue')
    plt.title(f'Potential U(h, v) with v = {v:.2f}', fontsize=14)
    plt.xlabel('h', fontsize=12)
    plt.ylabel('U(h, v)', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.ylim(0, 1)  # Adjust y-axis limits as needed
    plt.show()

# Create slider for v parameter
interact(
    plot_potential,
    v=FloatSlider(
        value=0.5,
        min=0,
        max=2,
        step=0.1,
        description='v value:',
        continuous_update=False
    )
);
```

```python
y0, v0 = 3, 0.01
t_span = (0,260000)
sol = solve_ivp(ode_sistema, t_span, [y0, v0], method='RK45', dense_output=True)  # Solución numérica

t = np.linspace(t_span[0],t_span[1], t_span[1]*20)

y, vel = sol.sol(t)[0], sol.sol(t)[1]  # Guardamos la solución numérica para el campo y su derivada.

potencial = 0.5*(1 + np.exp(-2*y))**(-2)  
damping = 3*vel*np.sqrt( potencial  + vel**2)
forcing = np.exp(-2*y) * (1 + np.exp(-2*y))**(-3)
acc = -damping - forcing # Obtenemos la solución numérica para la segunda derivada del campo.
```

```python
H = np.sqrt(0.5*vel**2 + potencial)/np.sqrt(6) 
dH_dt = np.gradient(H, t) 
epsilon = -dH_dt/H**2

idx1 = np.argmin(epsilon >= 1) 
idx2 = np.argmax(epsilon[idx1:] >= 1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, np.abs(damping), 'r-', label="Damping term")
plt.plot(t, np.abs(acc), 'b-', label=r"$\ddot\chi(\tilde{t})$")
plt.plot(t, np.abs(forcing), 'g-', label="Forcing term")
plt.title('Numerical solutions for: Damping term, Forcing term, and |y\'\'(t)|')
plt.axvline(t[idx2], color='green', linestyle='--', label=r"$\epsilon = 1$")
plt.xlim(0,t_span[1])
plt.ylim(-0.05, 0.3)  
plt.legend()
plt.xlabel(r"$t$", fontsize=16)
plt.grid()
plt.show()
```
